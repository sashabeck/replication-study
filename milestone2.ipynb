## Replication Study: Extralegal Collective Violence

This notebook will attempt to replicate a graph from the article "Extralegal Collective Violence in the United States, 1783–1865" published in the Journal of Digital History. The figure I replicated visualizes the number of incidents of extralegal collective violence resulting in fatalities per year between 1783 and 1865. Since the article provided the csv file used for the graph, I recreated the same graph using Python (Pandas and Matplotlib) to confirm the pattern that the author, Patrick Hoehne, reported can be reproduced.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Incidents_Year.csv")
data.head()
```

## Dataset

This dataset contains two variables:
1) Year: the year of recorded incidents  
2) Incidents: the number of extralegal collective violence incidents resulting in fatalities in that year

The dataset spans from 1783 to 1865.

```python
# resize graph so it is readable
plt.figure(figsize=(20,8))
plt.plot(data["Year"], data["Incidents"])

plt.title("figure 1: Incidents of Extralegal Collective Violence Resulting in Fatalities, 1783–1865")
plt.xlabel("Year")
plt.ylabel("Extralegal Collective Incidents Resulting in Fatalities")

# rotates labels on x-axis
plt.xticks(data["Year"], rotation=90)

plt.show()
```

## Replication Analysis

To replicate the author's visualization, I loaded the dataset Incidents_Year.csv from the article "Extralegal Collective Violence in the United States, 1783–1865" using Pandas. I then plotted the number of incidents per year using Matplotlib so that the resulting graph closely resembles the figure shown in the article.

Both my replicated graph and the article's graph display relatively low numbers of incidents during the late eighteenth and early nineteenth centuries, followed by a sharp increase in the 1850s. The spike in the late 1850s is very visible in both graphs, indicating a period of increased extralegal violence.

Any small differences between the graphs may result from visualization choices or matlab formatting rather than differences in the underlying data. For example, the article uses a wider plotting canvas and slightly different axis formatting. However, I did my best to include every year like the original graph did, not just every 5-10 years, even though there is some crowding on the x-axis. I also copied the x and y-axis and title wording exactly to duplicate the graph as accurately as possible.

Overall, the replicated graph supports the author's argument that extralegal collective violence increased significantly in the decades leading up to the Civil War. Using the data and code that the article included, this replication confirms that the visualization accurately reflects the underlying dataset. Additionally, by recreating the graph, I've discovered several methodological decisions the author made while creating his visualizations. As I had to purposely write code to ensure every year between 1784 and 1865 was displayed, I realized his concious decision of aggregating the incidents by individual year rather than by decade, which produced a more detailed view of fluctuations over time. The dataset also includes years with zero recorded incidents, which preserves continuity in the timeline rather than filtering them out. These choices influence how the historical trend is interpreted and make the rise in incidents during the 1850s more visible.
